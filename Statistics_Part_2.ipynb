```python
#Question 1 What is hypothesis testing in statistics?

#Answer Hypothesis testing is a way to use data to make a decision or draw a conclusion about something in the real world. Imagine you have a guess or assumption about a large group (called a population), but you can’t check every single person or item—so you take a smaller group (a sample) and test your idea using that.
```

```python
#Question2 What is the null hypothesis, and how does it differ from the alternative hypothesis?

#Answer The null hypothesis is like the “default” or “no-change” assumption. It says: “Nothing is happening” or “There’s no difference or effect.

# On the other hand, the alternative hypothesis (H₁) is saying, "Wait, something is different here!" It’s what you want to prove. In the same medicine example, the alternative hypothesis would be: "This new medicine works better than the old one."
```

```python
#Question3  What is the significance level in hypothesis testing, and why is it important?

#Answer The significance level (α) in hypothesis testing is a threshold used to decide whether to reject the null hypothesis. It's typically set before conducting the test, and it represents the maximum probability of making a Type I error, which occurs when the null hypothesis is rejected even though it is true.

#  Why is it Important?
# If your p-value (the result from your test) is less than α, you reject the null hypothesis.
# If the p-value is higher, you don’t reject the null.
```

```python
#Question4 What does a P-value represent in hypothesis testing?

#Answer In hypothesis testing, the p-value is a numerical measure that helps you determine the strength of the evidence against the null hypothesis (H₀). It quantifies the probability of obtaining your sample data (or something more extreme) if the null hypothesis were true.
```

```python
#Question5 How do you interpret the P-value in hypothesis testing?

#Answer The p-value helps you decide whether the results of your test are statistically significant or not. You interpret it by comparing it to a pre-decided significance level (usually 0.05).
```

```python
#Question6 What are Type 1 and Type 2 errors in hypothesis testing?

#Answer When we make decisions based on sample data, there’s always a chance we might make a mistake. These mistakes are called Type I and Type II errors.

# Type I	False alarm (rejecting H₀ when it’s true)
# Type II	Missed detection (not rejecting H₀ when it’s false)
```

```python
#Question7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

#Answer The one-tailed and two-tailed tests are variations of hypothesis testing, and the difference lies in how they assess the direction of the effect or difference in the data.

# One-Tailed Test
# Direction :- Tests one specific direction
# Staistical Power :- Higher (for specific direction)
# Appacability :- When direction matters

# Two-Tailed Test
# Direction :- Tests both directions
# Staistical Power :- Lower
# Appacability :- When direction doesn't matters
```

```python
#Question8 What is the Z-test, and when is it used in hypothesis testing?

#Answer A Z-test is a statistical test used to determine whether two population means are different when the population standard deviation is known and the sample size is reasonably large.You use the Z-distribution (standard normal distribution) to calculate the test statistic.

```

```python
#Question9 How do you calculate the Z-score, and what does it represent in hypothesis testing?

#Answer The Z-score is a statistical measure that indicates how many standard deviations an individual data point is from the mean of a dataset. It helps you understand how unusual or typical a value is compared to the distribution of the data.

# How to Calculate the Z-score: The formula is: 𝑍 = ( 𝑋 − 𝜇 ) 𝜎

# Where:  X ˉ = Sample mean
# μ = Population mean (under the null hypothesis)
# σ = Population standard deviation
# n = Sample size

# The Z-score tells you how many standard errors away your sample mean is from the population mean.

# It measures how unusual or extreme your sample result is if the null hypothesis were true.

# It helps you understand whether the difference between your sample and the population is statistically significant or just due to random chance.

```

```python
#Question10  What is the T-distribution, and when should it be used instead of the normal distribution?

#Answer The T-distribution (also called Student’s t-distribution) is a probability distribution similar to the normal distribution, but it’s wider and has thicker tails. It was designed to work better with small sample sizes and unknown population standard deviation.

# Use the T-distribution when:

#1 Your sample size is small (usually less than 30).

#2 You don’t know the population standard deviation (σ).

#3 The data is approximately normally distributed, or the sample size is large enough for it to behave normally.




```

```python
#Question11 What is the difference between a Z-test and a T-test?

#Answer Both tests are used to compare means and test hypotheses—but the main difference lies in sample size and what you know about the population.

# Z-test:
 # Use it when:

# The sample size is large (usually n > 30).

# The population standard deviation (σ) is known.

# Follows the Z (normal) distribution.

# Gives more precise results when there's a lot of data and less uncertainty.


#  T-test:
# Use it when:

# The sample size is small (usually n < 30).

# The population standard deviation is unknown.

# Follows the T-distribution.

# It adjusts for the extra uncertainty that comes with small samples by having thicker tails.
```

```python
#Question12 What is the T-test, and how is it used in hypothesis testing?

#Answer The T-test is a statistical method used to compare the means of two groups or a sample mean with a population mean to determine if there is a significant difference between them. It is especially useful when the population variance is unknown and the sample size is small ( 𝑛 < 30 ).

# State the Hypotheses:

# Null Hypothesis (H₀): There is no difference between the means (e.g., the two groups are the same).

# Alternative Hypothesis (H₁): There is a difference between the means.

# Calculate the T-statistic:
 # Formula:𝑇=(𝑋−𝜇)𝑠/𝑛

 # Determine the Critical Value or P-value:
# Use the T-distribution table (or software) to find the critical value for a chosen significance level ( 𝛼 ), based on degrees of freedom ( 𝑑 𝑓 = 𝑛 − 1 ).
# Alternatively, calculate the p-value for the observed T-statistic.

# Decision Rule:
# If the absolute value of the T-statistic is greater than the critical value, or the p-value is less than 𝛼 , reject H₀.
# Otherwise, fail to reject H₀.
```

```python
#Question13 What is the relationship between Z-test and T-test in hypothesis testing?

#Answer The Z-test and T-test both test hypotheses about means, but the key differences are:

# Z-test: Used for large samples (n > 30) and when the population standard deviation is known.

# T-test: Used for small samples (n < 30) and when the population standard deviation is unknown.
```

```python
#Question14 What is a confidence interval, and how is it used to interpret statistical results?

#Answer A confidence interval (CI) is a range of values, derived from sample data, that is used to estimate a population parameter (like a population mean). It provides an interval where we are fairly confident the true parameter lies, given the sample data.

# How It’s Used:

# Statistical Inference: It helps in decision-making. If the interval doesn’t include a certain value (e.g., zero or no effect), we might conclude the effect is statistically significant.
# Comparison: You can use CIs to compare the means of different groups and check if their intervals overlap or not.
```

```python
#Question15 What is the margin of error, and how does it affect the confidence interval?

#Answer The margin of error (MoE) is the amount by which you expect your sample estimate to differ from the true population value. It represents the uncertainty in the estimate and is used to calculate the confidence interval (CI).

#Effect on Confidence Interval:

# A larger margin of error results in a wider confidence interval, indicating more uncertainty about the estimate.
# A smaller margin of error leads to a narrower confidence interval, suggesting more precision in the estimate.
```

```python
#Question16 How is Bayes' Theorem used in statistics, and what is its significance?

#Answer Bayes' Theorem is a fundamental concept in statistics used to update the probability of a hypothesis based on new evidence. It combines prior knowledge (what we know before) and new data (what we observe) to calculate the posterior probability, which is the updated probability of a hypothesis being true.

# Significance in Statistics:

#1 Updating Probabilities: Bayes' Theorem allows us to update our beliefs or assumptions about a hypothesis as new data becomes available.
#2 Real-World Application: It’s widely used in decision-making, medical diagnostics, spam email filtering, and predictive models, where prior knowledge and new evidence need to be combined.
#3 Flexibility: It doesn't require a fixed sample size, and it works well in situations where we don’t have complete information.
```

```python
#Question17 What is the Chi-square distribution, and when is it used?

#Answer The Chi-square (χ²) distribution is a statistical distribution used primarily in hypothesis testing and in assessing the goodness of fit between observed and expected data. It's a special case of the Gamma distribution and is used when you're dealing with categorical data.

# When Is It Used?
#1 Goodness of Fit Test:
# To determine if the observed distribution of data fits an expected distribution. For example, checking if a dice is fair by comparing the observed frequencies of rolls to expected frequencies.

#2 Test for Independence:
# To assess if two categorical variables are independent. This is often used in contingency tables (like a survey comparing gender vs. choice of product).

#3 Homogeneity Test:
# To compare whether different samples have the same distribution of categorical data. It’s often used when comparing groups in studies, like testing if preferences for a product are the same across different regions.
```

```python
#Question18  What is the Chi-square goodness of fit test, and how is it applied?

#Answer The Chi-square goodness of fit test is a statistical test used to determine whether an observed distribution matches a theoretical or expected distribution. It's widely applied in categorical data analysis.

# How It Applied:

#1 State hypotheses: Null (H₀): Data fits the expected distribution. Alternative (H₁): Data does not fit.
#2 Calculate statistic: Use 𝜒 2 = ∑ ( 𝑂 𝑖 − 𝐸 𝑖 ) 2 𝐸 𝑖 , where 𝑂 𝑖 is observed, 𝐸 𝑖 is expected frequency.
#3 Compare values: Determine degrees of freedom ( 𝑑 𝑓 = 𝑘 − 1 ) and critical value from the Chi-square table.
#4 Decision: If 𝜒 2 > critical value, reject H₀. Otherwise, fail to reject H₀.
```

```python
#Question19  What is the F-distribution, and when is it used in hypothesis testing?

#Answer The F-distribution is a continuous probability distribution that arises frequently in statistical analysis, especially when comparing variances between two groups or testing the overall significance in regression analysis.

# When is the F-distribution used in hypothesis testing?

#1 Analysis of Variance (ANOVA):

# To test if there are significant differences among group means.

# Example: Comparing test scores among students from different school

#2 Variance Ratio Test:

# To test whether two populations have equal variances.

# Often used before choosing between a T-test (equal variances assumed) and Welch’s T-test (unequal variances).

#3 Regression Analysis:

# To test the overall significance of a regression model.

# It helps check if at least one predictor variable is significantly related to the outcome.
```

```python
#Question1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?

import scipy.stats as stats

# Sample data
sample_mean = 980       # mean of sample
population_mean = 1000  # known population mean
std_dev = 30            # known population standard deviation
n = 50                  # sample size
alpha = 0.05            # significance level

# Calculate Z-score
z_score = (sample_mean - population_mean) / (std_dev / (n ** 0.5))

# Calculate p-value for two-tailed test
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Output results
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the sample mean and the population mean.")
```

```python
#Question3 Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
from scipy import stats

# Step 1: Simulate sample data
np.random.seed(0)
sample_data = np.random.normal(loc=98, scale=10, size=40)  # mean=98, std=10, n=40

# Step 2: Known population parameters
population_mean = 100
population_std = 10  # Assumed known

# Step 3: Calculate sample mean and sample size
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Step 4: Calculate Z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Step 5: Calculate p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Step 6: Output results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population mean.")
```

```python
#Question4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample_mean, population_mean, population_std, n, alpha=0.05):
    """
    Performs a two-tailed Z-test and visualizes the decision region.

    Parameters:
    sample_mean (float): Mean of the sample data.
    population_mean (float): Mean of the population.
    population_std (float): Standard deviation of the population.
    n (int): Sample size.
    alpha (float): Significance level (default is 0.05).
    """
    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Calculate critical values for two-tailed test
    critical_value = norm.ppf(1 - alpha / 2)  # Upper critical value
    lower_critical_value = -critical_value   # Lower critical value

    # Determine p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Print results
    print(f"Z-score: {z_score}")
    print(f"Critical Values: {lower_critical_value:.2f}, {critical_value:.2f}")
    print(f"P-value: {p_value:.4f}")
    if abs(z_score) > critical_value:
        print("Reject the null hypothesis (statistically significant).")
    else:
        print("Fail to reject the null hypothesis (not statistically significant).")

    # Visualization
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, loc=0, scale=1)  # Standard normal distribution

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label="Standard Normal Distribution", color="blue")

    # Shade critical regions
    plt.fill_between(x, 0, y, where=(x > critical_value), color="orange", alpha=0.5, label="Reject H₀ (Right Tail)")
    plt.fill_between(x, 0, y, where=(x < lower_critical_value), color="orange", alpha=0.5, label="Reject H₀ (Left Tail)")

    # Shade acceptance region
    plt.fill_between(x, 0, y, where=(lower_critical_value < x) & (x < critical_value), color="green", alpha=0.5, label="Fail to Reject H₀")

    # Add Z-score to the plot
    plt.axvline(z_score, color="red", linestyle="--", label=f"Z-score = {z_score:.2f}")

    # Labels and legend
    plt.title("Two-Tailed Z-test Decision Region")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend(loc="upper right")
    plt.grid(True)
    plt.show()

# Example usage:
sample_mean = 110
population_mean = 100
population_std = 15
sample_size = 30
two_tailed_z_test(sample_mean, population_mean, population_std, sample_size)

```

```python
#Question5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(alpha=0.05, beta=0.2):
    """
    Visualizes Type I and Type II errors for hypothesis testing.

    Parameters:
    alpha (float): Significance level (probability of Type I error).
    beta (float): Probability of Type II error.
    """
    # Define distributions for null and alternative hypotheses
    x = np.linspace(-4, 4, 1000)
    null_dist = norm.pdf(x, loc=0, scale=1)  # Null hypothesis (centered at 0)
    alt_dist = norm.pdf(x, loc=2, scale=1)  # Alternative hypothesis (centered at 2)

    # Determine critical value for rejecting H0 (based on alpha)
    critical_value = norm.ppf(1 - alpha, loc=0, scale=1)

    # Plot the distributions
    plt.figure(figsize=(10, 6))
    plt.plot(x, null_dist, label="Null Hypothesis (H₀)", color="blue")
    plt.plot(x, alt_dist, label="Alternative Hypothesis (H₁)", color="red")

    # Highlight Type I error region
    plt.fill_between(x, 0, null_dist, where=(x > critical_value), color="orange", alpha=0.5, label="Type I Error (α)")

    # Highlight Type II error region
    plt.fill_between(x, 0, alt_dist, where=(x < critical_value), color="purple", alpha=0.5, label="Type II Error (β)")

    # Add annotations and legends
    plt.axvline(x=critical_value, color="black", linestyle="--", label="Critical Value")
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Test Statistic")
    plt.ylabel("Density")
    plt.legend(loc="upper right")
    plt.grid(True)
    plt.show()

# Example usage:
visualize_errors(alpha=0.05, beta=0.2)
```

```python
#Question6 Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy import stats

# Step 1: Simulate two independent samples
np.random.seed(42)

# Sample 1: Test scores from Group A (mean=70, std=10)
group_a = np.random.normal(loc=70, scale=10, size=30)

# Sample 2: Test scores from Group B (mean=75, std=10)
group_b = np.random.normal(loc=75, scale=10, size=30)

# Step 2: Perform independent two-sample t-test
t_statistic, p_value = stats.ttest_ind(group_a, group_b)

# Step 3: Output results
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the means of Group A and Group B.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the means of Group A and Group B.")
```

```python
#Question7 Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Step 1: Simulate paired sample data (before and after treatment)
np.random.seed(42)

# Simulate test scores before and after a treatment (e.g., a study program)
before_treatment = np.random.normal(loc=70, scale=10, size=30)  # Scores before treatment
after_treatment = before_treatment + np.random.normal(loc=5, scale=5, size=30)  # Scores after treatment

# Step 2: Perform paired sample t-test
t_statistic, p_value = stats.ttest_rel(before_treatment, after_treatment)

# Step 3: Output results
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between before and after treatment.")
else:
    print("Fail to reject the null hypothesis: No significant difference between before and after treatment.")

# Step 5: Visualization - Bar Chart of the Paired Samples
plt.figure(figsize=(8, 6))
x = np.arange(len(before_treatment))  # X-axis for individuals
width = 0.35  # Bar width

plt.bar(x - width/2, before_treatment, width, label='Before Treatment', color='blue', alpha=0.7)
plt.bar(x + width/2, after_treatment, width, label='After Treatment', color='green', alpha=0.7)

# Labels and title
plt.xlabel('Individuals')
plt.ylabel('Scores')
plt.title('Before vs After Treatment Scores')
plt.legend()
plt.xticks([])  # Hide x-ticks for clarity
plt.tight_layout()
plt.show()
```

```python
#Question9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(sample_data, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    sample_data (array-like): Sample data from which the confidence interval is to be calculated.
    confidence_level (float): Confidence level (default is 95%).

    Returns:
    tuple: Lower bound and upper bound of the confidence interval.
    """

    # Calculate sample mean and standard error
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
    sample_size = len(sample_data)

    # Calculate the standard error of the mean (SEM)
    standard_error = sample_std / np.sqrt(sample_size)

    # Get the critical value from the t-distribution (since we use sample data)
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

    # Calculate the margin of error
    margin_of_error = critical_value * standard_error

    # Calculate the confidence interval
    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

    return confidence_interval

# Example usage:
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample data

confidence_interval = calculate_confidence_interval(sample_data, confidence_level=0.95)
print(f"95% Confidence Interval: {confidence_interval}")
```

```python
#Question10 Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
import scipy.stats as stats

def calculate_margin_of_error(sample_data, confidence_level=0.95):
    """
    Calculate the margin of error for a given confidence level using sample data.

    Parameters:
    sample_data (array-like): Sample data from which the margin of error is to be calculated.
    confidence_level (float): Confidence level (default is 95%).

    Returns:
    float: The margin of error.
    """

    # Calculate sample mean and standard error
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (ddof=1 for sample)
    sample_size = len(sample_data)

    # Calculate the standard error of the mean (SEM)
    standard_error = sample_std / np.sqrt(sample_size)

    # Get the critical value from the t-distribution (since we use sample data)
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

    # Calculate the margin of error
    margin_of_error = critical_value * standard_error

    return margin_of_error

# Example usage:
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=100, scale=15, size=30)  # Simulated sample data

margin_of_error = calculate_margin_of_error(sample_data, confidence_level=0.95)
print(f"Margin of Error for 95% Confidence Level: {margin_of_error}")
```

```python
#Question11 Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayesian_inference(prior_H, likelihood_E_given_H, false_positive_rate, prior_not_H, likelihood_E_given_not_H):
    """
    Apply Bayes' Theorem to calculate the posterior probability of a hypothesis.

    Parameters:
    prior_H (float): The prior probability of the hypothesis (e.g., having the disease).
    likelihood_E_given_H (float): The likelihood of observing evidence given the hypothesis (e.g., the probability of a positive test result given the person has the disease).
    false_positive_rate (float): The false positive rate (the probability of a positive test result given the person does not have the disease).
    prior_not_H (float): The prior probability of the complement of the hypothesis (e.g., not having the disease).
    likelihood_E_given_not_H (float): The likelihood of observing evidence given the complement of the hypothesis (e.g., the probability of a positive test result given the person does not have the disease).

    Returns:
    float: The posterior probability P(H | E), i.e., the probability of the hypothesis given the evidence.
    """

    # Marginal likelihood P(E)
    marginal_likelihood_E = (likelihood_E_given_H * prior_H) + (likelihood_E_given_not_H * prior_not_H)

    # Bayes' Theorem: P(H | E) = (P(E | H) * P(H)) / P(E)
    posterior_H_given_E = (likelihood_E_given_H * prior_H) / marginal_likelihood_E

    return posterior_H_given_E

# Example usage for disease detection:
prior_H = 0.01  # 1% of the population has the disease
likelihood_E_given_H = 0.95  # 95% sensitivity (True Positive Rate)
false_positive_rate = 0.05  # 5% false positive rate
prior_not_H = 0.99  # 99% of the population does not have the disease
likelihood_E_given_not_H = false_positive_rate  # 5% false positive rate

# Calculate the posterior probability of having the disease given a positive test result
posterior_H_given_E = bayesian_inference(prior_H, likelihood_E_given_H, false_positive_rate, prior_not_H, likelihood_E_given_not_H)

print(f"The probability of having the disease given a positive test result is: {posterior_H_given_E:.4f}")
```

```python
#Question16  Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    """
    Performs an F-test to compare the variances of two samples.

    Parameters:
    sample1 (list or array): First sample data.
    sample2 (list or array): Second sample data.
    alpha (float): Significance level for the test (default is 0.05).

    Returns:
    None: Prints F-statistic, critical value, and result.
    """
    # Calculate sample variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    # Calculate F-statistic (larger variance divided by smaller variance)
    f_stat = var1 / var2 if var1 > var2 else var2 / var1

    # Degrees of freedom
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # Critical value for the F-distribution
    critical_value = f.ppf(1 - alpha, df1, df2)

    # Print results
    print(f"Sample 1 Variance: {var1:.4f}")
    print(f"Sample 2 Variance: {var2:.4f}")
    print(f"F-statistic: {f_stat:.4f}")
    print(f"Critical Value: {critical_value:.4f}")
    if f_stat > critical_value:
        print("Reject the null hypothesis: Variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in variances.")

# Generate random samples
np.random.seed(42)  # For reproducibility
sample1 = np.random.normal(10, 2, size=30)  # Mean=10, Std=2
sample2 = np.random.normal(15, 3, size=30)  # Mean=15, Std=3

# Perform the F-test
f_test(sample1, sample2)
```

```python
#Questiion20 Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

def two_way_anova():
    """
    Performs a two-way ANOVA test to analyze the interaction between two factors
    and visualizes the results using box plots.
    """
    # Example dataset: Two factors - 'Factor_A' and 'Factor_B' with dependent variable 'Value'
    data = {
        'Factor_A': np.repeat(['A1', 'A2'], 30),
        'Factor_B': np.tile(np.repeat(['B1', 'B2', 'B3'], 10), 2),
        'Value': np.random.normal(loc=10, scale=2, size=60) + np.tile([0, 1, -1], 20)
    }
    df = pd.DataFrame(data)

    # Perform two-way ANOVA using statsmodels
    model = ols('Value ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=df).fit()
    anova_results = anova_lm(model)

    # Print ANOVA results
    print("Two-way ANOVA Results:\n")
    print(anova_results)

    # Visualization: Box plots for interaction between factors
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Factor_B', y='Value', hue='Factor_A', data=df, palette="Set2")
    plt.title("Box Plot of Values by Factor Interaction")
    plt.xlabel("Factor B")
    plt.ylabel("Dependent Variable (Value)")
    plt.legend(title="Factor A")
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.show()

# Run the two-way ANOVA
two_way_anova()
```

```python
#Question24  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

import numpy as np
from scipy.stats import chi2

def variance_hypothesis_test(sample, hypothesized_variance, alpha=0.05):
    """
    Performs a hypothesis test for population variance using the Chi-square distribution.

    Parameters:
    sample (list or array): Sample data.
    hypothesized_variance (float): The hypothesized population variance.
    alpha (float): Significance level (default is 0.05).

    Returns:
    None: Prints the Chi-square statistic, critical values, and interpretation.
    """
    # Calculate sample variance
    sample_variance = np.var(sample, ddof=1)
    sample_size = len(sample)

    # Calculate Chi-square statistic
    chi_stat = (sample_size - 1) * sample_variance / hypothesized_variance

    # Determine critical values for two-tailed test
    lower_critical = chi2.ppf(alpha / 2, df=sample_size - 1)
    upper_critical = chi2.ppf(1 - alpha / 2, df=sample_size - 1)

    # Print results
    print("Hypothesis Test for Population Variance:")
    print(f"Sample Variance: {sample_variance:.4f}")
    print(f"Chi-square Statistic: {chi_stat:.4f}")
    print(f"Critical Values: Lower = {lower_critical:.4f}, Upper = {upper_critical:.4f}")

    # Decision
    if chi_stat < lower_critical or chi_stat > upper_critical:
        print("Reject the null hypothesis: The population variance is significantly different from the hypothesized variance.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in the population variance.")

# Example usage
np.random.seed(42)
sample_data = np.random.normal(10, 2, size=30)  # Sample data: Mean=10, StdDev=2
hypothesized_variance = 4  # Hypothesized variance

variance_hypothesis_test(sample_data, hypothesized_variance)
```

```python
#Question25 Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

import numpy as np
from statsmodels.stats.proportion import proportions_ztest

def z_test_for_proportions(successes, observations, alpha=0.05):
    """
    Performs a Z-test for comparing proportions between two groups.

    Parameters:
    successes (list): Number of successes in each group (e.g., [success_group1, success_group2]).
    observations (list): Total observations in each group (e.g., [n_group1, n_group2]).
    alpha (float): Significance level for the test (default is 0.05).

    Returns:
    None: Prints the Z-statistic, p-value, and interpretation.
    """
    # Perform the Z-test
    z_stat, p_value = proportions_ztest(count=successes, nobs=observations)

    # Print results
    print(f"Proportion Z-test Results:")
    print(f"Z-statistic: {z_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret the results
    if p_value < alpha:
        print("Reject the null hypothesis: The proportions are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference in proportions.")

# Example usage
successes = [50, 30]  # Number of successes in group1 and group2
observations = [100, 80]  # Total observations in group1 and group2

z_test_for_proportions(successes, observations)
```

```python
#Question26 Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to perform F-test for variances
def f_test_for_variances(data1, data2, alpha=0.05):
    # Calculate sample variances
    var1 = np.var(data1, ddof=1)  # Sample variance of data1
    var2 = np.var(data2, ddof=1)  # Sample variance of data2

    # Calculate the F-statistic (larger variance / smaller variance)
    f_statistic = var1 / var2 if var1 > var2 else var2 / var1

    # Degrees of freedom
    df1 = len(data1) - 1  # Degrees of freedom for data1
    df2 = len(data2) - 1  # Degrees of freedom for data2

    # Calculate the p-value for the F-test
    p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

    # Print results
    print(f"F-statistic: {f_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Hypothesis test decision
    if p_value < alpha:
        print("Reject the null hypothesis: The variances are significantly different.")
    else:
        print("Fail to reject the null hypothesis: The variances are not significantly different.")

    # Visualize the F-distribution and the calculated F-statistic
    x = np.linspace(0, 5, 1000)
    y = stats.f.pdf(x, df1, df2)

    plt.plot(x, y, label='F-distribution', color='blue')
    plt.axvline(f_statistic, color='red', linestyle='--', label=f'F-statistic = {f_statistic:.4f}')
    plt.title(f"F-distribution with df1={df1} and df2={df2}")
    plt.xlabel("F-value")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

# Simulating two random datasets (e.g., sample data from two groups)
np.random.seed(42)
data1 = np.random.normal(50, 10, 100)  # Mean=50, SD=10, N=100
data2 = np.random.normal(50, 15, 100)  # Mean=50, SD=15, N=100

# Perform the F-test for variances
f_test_for_variances(data1, data2, alpha=0.05)
```

```python
#Question27  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to perform Chi-square goodness of fit test
def chi_square_goodness_of_fit(observed, expected, alpha=0.05):
    # Calculate Chi-square statistic
    chi_square_stat = np.sum((observed - expected) ** 2 / expected)

    # Degrees of freedom
    df = len(observed) - 1

    # Calculate the p-value
    p_value = 1 - stats.chi2.cdf(chi_square_stat, df)

    # Print the results
    print(f"Chi-square statistic: {chi_square_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Hypothesis test decision
    if p_value < alpha:
        print("Reject the null hypothesis: The observed frequencies do not follow the expected distribution.")
    else:
        print("Fail to reject the null hypothesis: The observed frequencies follow the expected distribution.")

    # Visualizing the Chi-square distribution and the calculated statistic
    x = np.linspace(0, 20, 1000)
    y = stats.chi2.pdf(x, df)

    plt.plot(x, y, label=f'Chi-square distribution (df={df})', color='blue')
    plt.axvline(chi_square_stat, color='red', linestyle='--', label=f'Chi-square statistic = {chi_square_stat:.4f}')
    plt.title("Chi-square Distribution and Test Statistic")
    plt.xlabel("Chi-square value")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

# Simulating observed data (e.g., number of occurrences in 6 categories)
observed = np.array([20, 30, 25, 35, 40, 50])

# Expected frequencies (assuming an equal distribution across all categories)
expected = np.array([33, 33, 33, 33, 33, 33])

# Perform the Chi-square goodness of fit test
chi_square_goodness_of_fit(observed, expected, alpha=0.05)
```
