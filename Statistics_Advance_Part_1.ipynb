```python
#Question16  How do you calculate a Z-score, and what does it represent?

#Answer The Z-score (or standard score) is calculated using: ùëç = ùëã ‚àí ùúá ùúé Z= œÉ X‚àíŒº ‚Äã Where: ùëã X = observed value ùúá Œº = population mean ùúé œÉ = population standard deviation For sample data, the formula is: ùëç = ùëã ‚àí ùëã Àâ ùë† Z= s X‚àí X Àâ ‚Äã Where ùëã Àâ X Àâ is the sample mean and ùë† s is the sample standard deviation.

# What Does It Represent?
# A Z-score tells how many standard deviations an observation is from the mean.

# Positive Z-score ‚Üí Value is above the mean.

# Negative Z-score ‚Üí Value is below the mean.

# Z = 0 ‚Üí Value is exactly at the mean.
```

```python
#Question17 > What are point estimates and interval estimates in statistics?

#Answer Point Estimate vs. Interval Estimate

# Point Estimate ‚Üí A single value used to estimate a population parameter.
# Example: Sample mean ( ùëã Àâ X Àâ ) estimates population mean ( ùúá Œº).

# Interval Estimate ‚Üí A range of values likely to contain the population parameter, with a given confidence level.
# Example: Confidence Interval (CI), such as "The mean height is 170 cm ¬± 5 cm with 95% confidence."
```

```python
#Question18 What is the significance of confidence intervals in statistical analysis?

#Answer Significance of Confidence Intervals (CIs) in Statistical Analysis

# Measures Uncertainty ‚Üí Provides a range instead of a single estimate, accounting for sampling variability.

# Indicates Reliability ‚Üí A 95% CI means if we repeat the study many times, 95% of the intervals will contain the true population parameter.

# Helps Decision-Making ‚Üí If a confidence interval for a mean difference includes zero, the difference may not be statistically significant.

# More Informative than Point Estimates ‚Üí Shows both the estimate and its precision, unlike a single-value estimate.
```

```python
#Question19 What is the relationship between a Z-score and a confidence interval?

#Answer A Z-score is used to determine the critical value for a given confidence level in a confidence interval (CI) calculation.

# Key Relationship:
# A higher confidence level requires a larger Z-score, leading to a wider CI (more certainty).

# A lower confidence level has a smaller Z-score, resulting in a narrower CI (less certainty).
```

```python
#Question4 > Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def generate_discrete_uniform(n, a, b):
    """Generate a discrete uniform distribution."""
    return np.random.randint(a, b + 1, size=n)

def bernoulli_pdf(x, p):
    """Calculate the probability distribution function (PDF) of a Bernoulli distribution."""
    return p if x == 1 else (1 - p) if x == 0 else 0

def simulate_binomial(n, p, size):
    """Simulate a binomial distribution."""
    return np.random.binomial(n, p, size)

# Parameters for discrete uniform distribution
n_samples = 10000  # Number of samples
a, b = 1, 6  # Range of discrete values (like a fair die)

# Generate discrete uniform data
data = generate_discrete_uniform(n_samples, a, b)

# Compute PMF
values, counts = np.unique(data, return_counts=True)
probabilities = counts / n_samples

# Plot PMF
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(values, probabilities, width=0.6, color='skyblue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Discrete Uniform Distribution PMF')
plt.xticks(values)
plt.ylim(0, max(probabilities) + 0.05)

# Parameters for binomial distribution
n_binomial = 10  # Number of trials
p_binomial = 0.5  # Probability of success
size_binomial = 10000  # Number of samples

# Generate binomial data
binomial_data = simulate_binomial(n_binomial, p_binomial, size_binomial)

# Plot histogram of binomial distribution
plt.subplot(1, 2, 2)
plt.hist(binomial_data, bins=range(n_binomial + 2), density=True, alpha=0.75, color='orange', edgecolor='black')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution Histogram')
plt.xticks(range(n_binomial + 1))
plt.ylim(0, max(np.bincount(binomial_data) / size_binomial) + 0.05)

plt.tight_layout()
plt.show()
```

```python
#Question6 Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
low = 1   # Minimum value (inclusive)
high = 6  # Maximum value (inclusive)

# Generate the values and their probabilities
values = np.arange(low, high + 1)
probabilities = np.full(len(values), 1 / len(values))  # Equal probability for each value

# Calculate the CDF
cdf = np.cumsum(probabilities)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(values, cdf, where='post', color='blue', label="CDF")
plt.scatter(values, cdf, color='red')  # Highlight CDF points
plt.title("Cumulative Distribution Function (CDF) of Discrete Uniform Distribution")
plt.xlabel("Values")
plt.ylabel("Cumulative Probability")
plt.xticks(values)
plt.grid(alpha=0.5, linestyle='--')
plt.legend()
plt.show()
```

```python
#Question9 Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Function to calculate Z-scores from a dataset and plot them.
    Parameters:
        data (array-like): Input dataset (list or NumPy array).
    """
    # Convert to NumPy array if data is not already
    data = np.array(data)

    # Calculate the mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot the Z-scores
    plt.figure(figsize=(8, 5))
    plt.plot(data, z_scores, marker='o', color='blue', label='Z-scores')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.title("Z-Scores of the Dataset")
    plt.xlabel("Data Points")
    plt.ylabel("Z-Score")
    plt.grid(alpha=0.5, linestyle='--')
    plt.legend()
    plt.show()

    # Return Z-scores for reference
    return z_scores

# Example usage
data = [10, 12, 15, 18, 20, 21, 25]  # Sample dataset
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)
```

```python
#Question10 Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt

# Parameters for the uniform distribution (non-normal distribution)
low = 0   # Minimum value
high = 10 # Maximum value
population_size = 100000  # Size of the population

# Generate a non-normal population
population = np.random.uniform(low, high, population_size)

# Simulate the Central Limit Theorem
sample_sizes = [5, 10, 30, 50, 100]  # Various sample sizes
means = {}  # To store sample means for each sample size

for size in sample_sizes:
    sample_means = []
    for _ in range(1000):  # Number of samples to take
        sample = np.random.choice(population, size=size, replace=False)
        sample_means.append(np.mean(sample))
    means[size] = sample_means

# Plot the distribution of sample means for each sample size
plt.figure(figsize=(12, 8))
for size, sample_means in means.items():
    plt.hist(sample_means, bins=30, density=True, alpha=0.6, label=f'Sample size = {size}')

plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.5, linestyle='--')
plt.show()
```

```python
#Question12 Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)

import numpy as np
import matplotlib.pyplot as plt

def plot_standard_normal_distribution():
    """
    Function to calculate and plot the Standard Normal Distribution.
    Mean = 0, Standard Deviation = 1.
    """
    # Define the range of x values (3 standard deviations on either side of the mean)
    x = np.linspace(-4, 4, 1000)  # 1000 points between -4 and 4
    mean = 0
    std_dev = 1

    # Calculate the Probability Density Function (PDF)
    pdf = (1 / (np.sqrt(2 * np.pi) * std_dev)) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

    # Plot the standard normal distribution
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, color='blue', label='Standard Normal Distribution (mean=0, std=1)')
    plt.title("Standard Normal Distribution")
    plt.xlabel("Value")
    plt.ylabel("Probability Density")
    plt.grid(alpha=0.5, linestyle='--')
    plt.legend()
    plt.show()

# Call the function
plot_standard_normal_distribution()
```

```python
#Question14 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    """
    Calculate the Z-score for a given data point.
    Parameters:
        data_point (float): The value for which Z-score is to be calculated.
        mean (float): The mean of the dataset.
        std_dev (float): The standard deviation of the dataset.
    Returns:
        float: The Z-score.
    """
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_to_standard_normal(z_score):
    """
    Compare the Z-score to the standard normal distribution by calculating cumulative probability.
    Parameters:
        z_score (float): The Z-score to compare.
    Returns:
        float: The cumulative probability (area under the curve).
    """
    cumulative_probability = norm.cdf(z_score)  # Area to the left of the Z-score
    return cumulative_probability

# Example usage
data_point = 70
mean = 50
std_dev = 10

# Calculate Z-score
z_score = calculate_z_score(data_point, mean, std_dev)
print(f"Z-score for data point {data_point}: {z_score:.2f}")

# Compare Z-score to standard normal distribution
cumulative_probability = compare_to_standard_normal(z_score)
print(f"Cumulative probability for Z-score {z_score:.2f}: {cumulative_probability:.4f}")

# Visualization
import matplotlib.pyplot as plt
x = np.linspace(-4, 4, 1000)  # Z-scores from -4 to 4
pdf = norm.pdf(x)  # Probability density function of standard normal distribution

plt.figure(figsize=(8, 5))
plt.plot(x, pdf, color='blue', label='Standard Normal PDF')
plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score: {z_score:.2f}')
plt.fill_between(x, 0, pdf, where=(x <= z_score), color='skyblue', alpha=0.5, label='Cumulative Probability')
plt.title("Comparison of Z-score to Standard Normal Distribution")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.legend()
plt.grid(alpha=0.5, linestyle='--')
plt.show()
```

```python
#Question15  Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

def hypothesis_test_z(sample, population_mean, population_std_dev, alpha=0.05):
    """
    Perform hypothesis testing using Z-statistics.

    Parameters:
        sample (list or array): The sample dataset.
        population_mean (float): The mean of the population (null hypothesis mean).
        population_std_dev (float): The standard deviation of the population.
        alpha (float): Significance level (default: 0.05 for 5%).

    Returns:
        dict: A dictionary containing Z-statistic, p-value, and result of the test.
    """
    # Step 1: Calculate the sample mean and sample size
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Step 2: Calculate the Z-statistic
    z_stat = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    # Step 3: Calculate the p-value (for two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Step 4: Compare p-value with alpha to decide
    reject_null = p_value < alpha

    # Print the results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_stat:.2f}")
    print(f"p-value: {p_value:.4f}")
    if reject_null:
        print("Result: Reject the null hypothesis (H0).")
    else:
        print("Result: Fail to reject the null hypothesis (H0).")

    # Return results as a dictionary
    return {"z_stat": z_stat, "p_value": p_value, "reject_null": reject_null}

# Example usage
sample_data = [52, 48, 50, 47, 51, 49, 54, 48, 50, 46]  # Sample dataset
population_mean = 50  # Population mean (under H0)
population_std_dev = 3  # Population standard deviation
alpha = 0.05  # Significance level

# Perform hypothesis testing
result = hypothesis_test_z(sample_data, population_mean, population_std_dev, alpha)
```

```python
#Question16 Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy.stats import norm

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a dataset.
    Parameters:
        data (array-like): Dataset values (list or NumPy array).
        confidence_level (float): Desired confidence level (default: 95%).
    Returns:
        tuple: Lower and upper bounds of the confidence interval.
    """
    # Convert to NumPy array if not already
    data = np.array(data)

    # Step 1: Calculate the sample mean and standard deviation
    sample_mean = np.mean(data)
    sample_std_dev = np.std(data, ddof=1)  # Use ddof=1 for sample std deviation
    sample_size = len(data)

    # Step 2: Calculate the margin of error
    z_critical = norm.ppf((1 + confidence_level) / 2)  # Z-value for the desired confidence level
    margin_of_error = z_critical * (sample_std_dev / np.sqrt(sample_size))

    # Step 3: Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Margin of Error: ¬±{margin_of_error:.2f}")
    print(f"Confidence Interval ({confidence_level*100:.0f}%): [{lower_bound:.2f}, {upper_bound:.2f}]")

    return lower_bound, upper_bound

# Example usage
data = [45, 50, 55, 60, 65, 70, 75, 80]  # Sample dataset
confidence_level = 0.95  # 95% confidence level

# Calculate the confidence interval
lower, upper = calculate_confidence_interval(data, confidence_level)
```

```python
#Question19  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

def plot_poisson_cdf(lmbda, max_value):
    """
    Calculate and plot the CDF of a Poisson distribution.

    Parameters:
        lmbda (float): Mean of the Poisson distribution (Œª).
        max_value (int): Maximum value for the random variable.
    """
    # Generate x values (possible outcomes)
    x = np.arange(0, max_value + 1)

    # Calculate the CDF using scipy.stats
    cdf_values = poisson.cdf(x, mu=lmbda)

    # Plot the CDF
    plt.figure(figsize=(8, 5))
    plt.step(x, cdf_values, where='post', color='blue', label=f'Poisson CDF (Œª={lmbda})')
    plt.scatter(x, cdf_values, color='red', label='CDF Points')
    plt.title(f'Cumulative Distribution Function (CDF) of Poisson Distribution (Œª={lmbda})')
    plt.xlabel('Number of Events')
    plt.ylabel('Cumulative Probability')
    plt.grid(alpha=0.5, linestyle='--')
    plt.legend()
    plt.show()

    # Print interpretation for example values
    print(f"For example, P(X ‚â§ 3): {cdf_values[3]:.4f} (probability of up to 3 events)")
    return cdf_values

# Example usage
lambda_param = 5  # Mean number of events (Œª)
max_event_count = 15  # Max value for the random variable

cdf_values = plot_poisson_cdf(lambda_param, max_event_count)
```

```python
#Question21 Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(dataset1, dataset2):
    """
    Compare the standard deviations of two datasets and visualize the difference.
    Parameters:
        dataset1 (list or array): First dataset.
        dataset2 (list or array): Second dataset.
    """
    # Calculate the standard deviations
    std_dev1 = np.std(dataset1, ddof=1)  # Sample standard deviation
    std_dev2 = np.std(dataset2, ddof=1)

    # Print the results
    print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")

    # Visualize the difference
    datasets = ['Dataset 1', 'Dataset 2']
    std_devs = [std_dev1, std_dev2]

    plt.figure(figsize=(8, 5))
    plt.bar(datasets, std_devs, color=['skyblue', 'lightgreen'], edgecolor='black', alpha=0.7)
    plt.title("Comparison of Standard Deviations")
    plt.ylabel("Standard Deviation")
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.show()

# Example datasets
dataset1 = [10, 15, 20, 25, 30, 35]
dataset2 = [5, 12, 18, 40, 55, 60]

# Compare and visualize the standard deviations
compare_standard_deviations(dataset1, dataset2)
```

```python
#Question23 Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Perform Z-score normalization on a dataset and visualize its transformation.
    Parameters:
        data (array-like): The input dataset (list or NumPy array).
    Returns:
        array: Z-score normalized data.
    """
    # Convert to NumPy array if not already
    data = np.array(data)

    # Calculate mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Perform Z-score normalization
    z_scores = (data - mean) / std_dev

    # Visualization: Original vs Normalized Data
    plt.figure(figsize=(12, 6))

    # Histogram of Original Data
    plt.subplot(1, 2, 1)
    plt.hist(data, bins=20, color='lightblue', edgecolor='black', alpha=0.7)
    plt.title("Original Dataset")
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.grid(alpha=0.5, linestyle="--")

    # Histogram of Z-score Normalized Data
    plt.subplot(1, 2, 2)
    plt.hist(z_scores, bins=20, color='lightgreen', edgecolor='black', alpha=0.7)
    plt.title("Z-score Normalized Dataset")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.grid(alpha=0.5, linestyle="--")

    # Display the plots
    plt.tight_layout()
    plt.show()

    return z_scores

# Example usage
data = [120, 130, 125, 115, 140, 135, 145, 150, 110, 105]  # Sample dataset
normalized_data = z_score_normalization(data)
```

```python
#Question24 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    """
    Calculate skewness and kurtosis for a given dataset.
    Parameters:
        data (array-like): Input dataset (list or NumPy array).
    Returns:
        tuple: Skewness and kurtosis values.
    """
    # Calculate skewness
    data_skewness = skew(data)

    # Calculate kurtosis
    data_kurtosis = kurtosis(data, fisher=True)  # Fisher=True gives excess kurtosis

    # Print the results
    print(f"Skewness: {data_skewness:.2f}")
    print(f"Kurtosis (Excess): {data_kurtosis:.2f}")

    return data_skewness, data_kurtosis

# Generate a dataset from a normal distribution
mean = 0       # Mean of the distribution
std_dev = 1    # Standard deviation
size = 1000    # Number of samples

# Generate normal data
data = np.random.normal(mean, std_dev, size)

# Calculate and display skewness and kurtosis
skewness, kurt = calculate_skewness_kurtosis(data)
```
